```python
import tensorflow as tf
print(tf.__version__)
from tensorflow import keras
fashion= tf.keras.datasets.fashion_mnist
import numpy as np
```

```python
(training_images,training_labels),(test_images,test_labels)=fashion.load_data()
```

```python
class_name =['t_shirt','trouser','pullover','dress','coat','sandel','shirt','sneaker','bag','ankle boot']
print("training_images =",training_images.shape)
print("training_labels =",len(training_labels))
print("test_images =",test_images.shape)
print("test_labels=", len(test_labels))
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(training_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

to make normalization


```python
training_images = training_images/255.0
test_images= test_images/255.0
```

get & process data

```python
plt.figure(figsize=(10,10))
for i in range (30):
  plt.subplot(6,6,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(training_images[i],cmap=plt.cm.binary)
  plt.xlabel(class_name[training_labels[i]])
plt.show()
```

building neural network

```python
from keras.engine.training import optimizer
model = keras.Sequential(
    [keras.layers.Flatten(input_shape=(28,28)),
     keras.layers.Dense(128,activation='relu'),
     keras.layers.Dense(10,activation='softmax')])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuarcy'])
model.fit(training_images,training_labels,epochs=30)
```

evaluate model

```python
test_loss,test_acc= model.evaluate(test_images, test_labels,verbose=2)
print('\n test accuarcy',test_acc)
```

predicitions

```python
predict =model.predict(test_images)
np.argmax(predict([0]))
```
